# matplotlib으로 데이터 시각화
## 데이터 시각화
- 데이터 시각화는 점이나 선, 막대 그래프의 시각적 이미지를 사용하여 데이터를 화면에 표시함 
- 데이터 시각화는 시각적으로 보이는 데이터를 직관적으로 이해할 수 있기 때문에 사용함 
- 데이터를 시각화 하기 위해서는 확률의 개념이 많이 사용되며, 
- 평균값, 가우시안 확률, 중간값, 산점도 등의 다양한 방법 사용 
- matplotlib, seaborn등의 다양한 라이브러리가 제공됨

```python
import matplotlib.pyplot as plt

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [67, 80, 257, 1686, 6505, 11865, 22105]

plt.plot(years, gdp, color='blue', marker='o', linestyle='solid')
plt.title("GDP per capita")
plt.ylabel("dollars")
plt.show()
```

# matplotlib
- matplotlib는 파이썬으로 도식화를 할 수 있는 라이브러리를 제공하며, 선, 막대, 산점도 등  다양한 시각화를 지원함 
- pyplot는 matlab과 같은 인터페이스를 제공하는 matplotlib의 모듈 
- 기본적으로 사용하는 함수 
    - matplotlib.pyplot.plot( data ……) 
    - matplotlib.pyplot.bar( data ……) 
    - matplotlib.pyplot.scatter( data ……) 
    - matplot.pyplot.pie( data ……) 
- 그리는 순서
    - plot->title->x, y label -> legend -> show

### 여러개의 선 그리기
- 리스트로 여러 개의 데이터를 선언한다. (x, y, z) 
- 여기에 각 차트를 설명하는 범례를 추가하면 다음과 같다. 
- plot.plt를 사용하며, 각기 다른 입력데이터, marker, linestyle, label 의 특성을 다르게 해서 선을 그릴 수 있다.

```python
## 한 차트에 복수의 선 그리기

# 연도 데이터
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]

# GDP 데이터
gdp1 = [67, 80, 257, 886, 6505, 11865, 22105]
gdp2 = [100, 200, 500, 700, 900, 1200, 1800]
gdp3 = [200, 400, 700, 1200, 8000, 15000, 30000]

# 첫 번째 GDP 그래프 (파란색, 원 마커)
plt.plot(years, gdp1, color='blue', marker='o', linestyle='solid', label='GDP1')

# 두 번째 GDP 그래프 (초록색, X 마커)
plt.plot(years, gdp2, color='green', marker='x', linestyle='solid', label='GDP2')

# 세 번째 GDP 그래프 (빨간색, 별 마커)
plt.plot(years, gdp3, color='red', marker='*', linestyle='solid', label='GDP3')
plt.title("GDP per capita")
plt.xlabel("Year")
plt.ylabel("Dollars")
plt.legend() #범례
plt.show()
```

### sine, cosline 그리기

```python
import math

x = []
y_sin = []
y_cos = []

for angle in range(360) :
    x.append(angle)
    y_sin.append(math.sin(math.radians(angle)))
    y_cos.append(math.cos(math.radians(angle)))

plt.plot(x,y_sin, label='Sine', color='blue')
plt.plot(x,y_cos, label='Cosine', color='red')

plt.title('Sine-Cosine Wave')
plt.legend()
plt.show()
```

### 막대형 차트
*xticks() 함수는 가로축의 범위의 눈금을 지정 

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [67, 80, 257, 1686, 6505, 11865, 22105]

plt.bar(range(len(years)), gdp)
plt.title('GDP per capita')
plt.ylabel("dollars")
plt.xticks(range(len(years)),years)
plt.show()
```

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]

# GDP 데이터
gdp1 = [67, 80, 257, 886, 6505, 11865, 22105]
gdp2 = [100, 200, 500, 700, 900, 1200, 1800]
gdp3 = [200, 400, 700, 1200, 8000, 15000, 30000]

years_range = range(len(years))

plt.bar(years_range, gdp1, width=0.25)
plt.bar(years_range, gdp2, width=0.25)
plt.bar(years_range, gdp3, width=0.25)

plt.title('GDP per capita')
plt.ylabel("dollars")
plt.xticks(range(len(years)), years) #years의 길이만큼 years를 x 축에 심기
plt.show()
```

```python
import numpy as np

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]

# GDP 데이터
gdp1 = [67, 80, 257, 886, 6505, 11865, 22105]
gdp2 = [100, 200, 500, 700, 900, 1200, 1800]
gdp3 = [200, 400, 700, 1200, 8000, 15000, 30000]

x_years_range = np.arange(len(years))

plt.bar(x_years_range, gdp1, width=0.25, label='gdp1')
plt.bar(x_years_range+0.3, gdp2, width=0.25, label='gdp2')
plt.bar(x_years_range+0.6, gdp3, width=0.25, label='gdp3')

plt.title("GDP per capita")
plt.ylabel("dollars")
plt.xticks(range(len(years)), years)
plt.legend()
plt.show()
```

### 산포도 그리기
- 산포도는 선이 연결되지 않고 scatter() 함수를 이용해서 그린다. 
- 데이터의 분포를 확인하는데 있어서 편리한 도식화 도구이다.

```python
datax = np.arange(10, 50) #10부터 50개 만큼의 데이터를 array 안에 넣음
datay = datax + 10*np.random.randn(40)
# np.random.randn(n): 평균 0, 표준편차 1을 따르는 표준 정규분포에서 n개의 난수를 생성
# np.random.randn(40) → 표준 정규분포에서 난수 40개 생성

plt.scatter(datax, datay)
plt.title('X data vs Y data')
plt.xlabel('X Scatter data')
plt.ylabel('y scatter data')
plt.show()
```

### 파이차트 그리기
- 파이차트는 데이터의 값에 따라서 원형 비율로 나누어져 있는 차트이다. 
- 파이차트는 비즈니스, 개인의 자산 등의 내용에 대해서 한눈에 쉽게 직관적으로 볼 수 있도록 한다. 

```python
stock_data = [20, 10, 15, 30, 25, 5]
stock_data_label = ['Apple', 'Persimon', 'Grape', 'Strawberry', 'Mango', 'Orange']

plt.pie(stock_data, labels=stock_data_label, autopct="%.2f") 
# autopct는 부채꼴 안에 표시될 숫자의 형식을 지정
plt.show()
```

```python
import pandas as pd

# 과일별 재고 데이터 (Series 형태)
stock_data = pd.Series([20, 10, 15, 30, 25, 5],\
                       index = ['Apple', 'Persimon', 'Grape', 'Strawberry', 'Mango', 'Orange'])

# 특정 항목을 강조하기 위한 explode 설정 (조각을 밖으로 빼내는 효과)
# 0.2, 0.1, 0.05는 원에서 더 튀어나오게 설정
explode = pd.Series([0.2, 0.1, 0.05, 0, 0, 0], index = stock_data.index)

# 각 조각의 비율(%)을 소수점 2자리까지 표시
plt.pie(stock_data, labels=stock_data.index, autopct='%.2f', explode=explode)
plt.show()
```

**결과**
- Apple (20%) → 20% 밖으로 이동 (0.2)
- Persimon (10%) → 10% 밖으로 이동 (0.1)
- Grape (15%) → 5% 밖으로 이동 (0.05)
- 나머지는 원래 위치에 유지됨.

```python
stock_data = pd.Series([20, 10, 15, 30, 25, 5],\
                       index = ['Apple', 'Persimon', 'Grape', 'Strawberry', 'Mango', 'Orange'])
wedgeproposition = {'width' : 0.6, 'linewidth' : 1, 'edgecolor' : 'black'}
explode = pd.Series([0.2, 0.1, 0.05, 0, 0, 0], index = stock_data.index)
plt.pie(stock_data, labels=stock_data.index, autopct='%.2f', explode=explode, shadow=True, wedgeprops=wedgeproposition, pctdistance=0.7)
plt.title('title', size=15)
plt.legend(loc = (1,0.5))
plt.show()
```

### 다중 히스토그램

```python
# 정규 분포의 평균(mu)과 표준편차(sigma) 설정
mu1, sigma1 = 10, 2
mu2, sigma2 = -6, 3

# 1. 표준 정규 분포 (평균 0, 표준편차 1)
standard_Gauss = np.random.randn(10000)
# 평균 0, 표준편차 1인 표준 정규 분포에서 난수 10,000개 생성

# 2. 평균 10, 표준편차 2인 정규 분포
Gauss1 = mu1 + sigma1 * np.random.randn(10000)

# 3. 평균 -6, 표준편차 3인 정규 분포
Gauss2 = mu2 + sigma2 * np.random.randn(10000)

plt.figure(figsize=(10,6))

# 표준 정규 분포 히스토그램 (반투명)
plt.hist(standard_Gauss, bins=50, alpha=0.4)
# alpha는 투명도를 의미합니다. 0.0에서 1.0 사이의 값을 갖습니다.

plt.hist(Gauss1, bins=50, alpha=0.4)
plt.hist(Gauss2, bins=50, alpha=0.4)
plt.show()
```
